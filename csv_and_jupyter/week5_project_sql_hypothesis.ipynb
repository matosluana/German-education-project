```python
# I will keep only the columns I need for my testing
data = data.drop(columns=['average_age',
       'inhabitants_up_to_17_years_in_%.',
       'inhabitants_from_18_to_64_years_in_%.','unemployment_density','social_benefits','employment_density',
       'inhabitants_from_65_years_in_%','population_development_(5-year_comparison)', 'households',
       'single-person_households_in_%', 'families_with_children_in_%','total_households', 'until_1000€', 'over_1000_until_3000€',
       'over_3000_until_5000€', 'over_5000€',
       'pupils_at_general_education_schools',])

data
```

```python
data =data[data['proportion_of_students_transferring_to_gymnasium'] !=0]
data
```

```python
#creating income bins

def income_bin(income):
    if income <=2000:
        return 'very low'
    elif income >2000 and income <=2750:
        return 'low'
    elif income >2750 and income <=3500:
        return 'medium'
    elif income >3500 and income <4250:
        return 'high'
    elif income >4250:
        return 'very high'
    

data['income_bin']= data['median_income_€'].apply(income_bin)
data
```

### Hypothesis testing only for the very high incomes

```python
# H0 = very high income acceptance rates <= other incomes acceptance rates 
# H1 = very high income acceptance rates > other incomes acceptance rates

data_very_high_income = data[data['income_bin'] == 'very high']
data_very_high_income
```

```python
data_without_very_high = data[data['income_bin'] != 'very high']
median_without_very_high = np.median(data_without_very_high['proportion_of_students_transferring_to_gymnasium'])
median_without_very_high
```

```python
stat, pval = ttest_1samp(data_very_high_income['proportion_of_students_transferring_to_gymnasium'], median_without_very_high)
```

```python
print('stat is  ', stat)
print('pvalue for the two-tailed test is ', pval)

print('pvalue for the one-tailed test is ', pval/2)
```

<b> H0 = higher income acceptance rates <= lower incomes acceptance rates</b>
 
Can be discarded, since pvalues <0.05 and stat is positive
                                                                      

<b> H1 = higher income acceptance rates > lower incomes acceptance rates</b>
    
Accepted because stat is positive

### Hypothesis testing for high incomes against low and medium incomes

```python
# H0 = high income acceptance rates <= low and medium incomes acceptance rates 
# H1 = high income acceptance rates acceptance rates > low and medium acceptance rates

data_high_income = data[data['income_bin'] == 'high']
data_high_income
```

```python
data_medium_low_income = data[data['income_bin'] != 'high']
data_medium_low_income = data_medium_low_income[data_medium_low_income['income_bin'] != 'high']

median_medium_low_income = np.median(data_medium_low_income['proportion_of_students_transferring_to_gymnasium'])
median_medium_low_income
```

```python
stat, pval = ttest_1samp(data_high_income['proportion_of_students_transferring_to_gymnasium'], median_medium_low_income)
```

```python
print('stat is  ', stat)
print('pvalue for the two-tailed test is ', pval)

print('pvalue for the one-tailed test is ', pval/2)
```

<b>H0 = high income acceptance rates <= low and medium incomes acceptance rates

<b> H1 = high income acceptance rates acceptance rates > low and medium acceptance rates</b>

H1 is accepted. There is a high probability that more students from neghbourhoods with higher income will be transferred to Gymnasium in comparison with lower income neighbourhoods.
